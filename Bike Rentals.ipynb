Let's add the column 'time_label', to split the rentals
by part of the day:<br>
morning, afternoon, evening, and night.

```python
def assign_label(num_hr):
    if (num_hr > 6) & (num_hr <= 12):
        rt = 1
    elif (num_hr > 12) & (num_hr <= 18):
        rt = 2
    elif (num_hr > 18) & (num_hr <= 24):
        rt = 3
    else:
        rt = 4
    return rt

bike_rentals['time_label'] = bike_rentals['hr'].apply(assign_label)
```

```python
rentals_by_day_part = bike_rentals.groupby('time_label')['cnt'].sum()
fig = plt.figure(figsize=(6,4))
ax = plt.axes()
ax.bar(['morning', 'afternoon', 'evening', 'night'], rentals_by_day_part.values)

# ax.set_yticklabels([])
# ax.get_yaxis().set_visible(False)
# plt.axis('off')
ax.set_title("Rentals by Part of the Day")
plt.show()
```

Bikes are rented more in the afternoon, then in the morning. <br>
Evening, meaning till midnight, has still a good level of rentals.<br>
Night, from midnight to six am, is clearly less frequent.

## Error Metric

As **error metric**, we should go on with the mean squared error (**MSE**).<br>
The 'squared' part punishes enough big errors.

```python
train = bike_rentals.sample(frac=0.8)
test = bike_rentals.loc[~bike_rentals.index.isin(train.index)]
```

We just built train and test datasets, <br>
train gets the 80% of bike_rentals, <br>
and test its 20%.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

columns = list(bike_rentals.columns)
remove_cols = ['cnt', 'casual', 'dteday', 'registered']
columns = [x for x in columns if x not in remove_cols]

lr = LinearRegression()
lr.fit(train[columns], train['cnt'])
predictions = lr.predict(test[columns])
# print(np.mean((predictions - test['cnt']) ** 2))
mean_squared_error(test['cnt'], predictions)
```

As we said before, MSE penalizes big mistakes. <br>
We have big mistakes here, because the MSE number is huge!

```python
predictions2 = lr.predict(train[columns])
mean_squared_error(train['cnt'], predictions2)
```

At least there is no overfitting here, <br>
but overfitting is not strictly a problem of Logistic Regression!

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(min_samples_leaf=2)
model.fit(train[columns], train['cnt'])
predictions_decision = model.predict(test[columns])
mean_squared_error(test['cnt'], predictions_decision)
```

```python
predictions_decision_train = model.predict(train[columns])
mean_squared_error(train['cnt'], predictions_decision_train)
```

When we talk about overfitting this is exactly what we talk about. <br>
On the training set, our prediction is almost perfect.<br>
On the other hand, the MSE on test set is<br>
still less than 1/6th what he got with Logistic Regression.<br>

Let's try to see what we get with an ensemble method, <br>
like Random Forest.

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(min_samples_leaf=2)
regressor.fit(train[columns], train['cnt'])
predictions_random = regressor.predict(test[columns])
mean_squared_error(test['cnt'], predictions_random)
```

```python
predictions_random_train = regressor.predict(train[columns])
mean_squared_error(train['cnt'], predictions_random_train)
```

The result is way better, but we can still experience<br>
some overfitting. Let's try to change some parameter.

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(min_samples_leaf=5)
regressor.fit(train[columns], train['cnt'])
predictions_random = regressor.predict(test[columns])
mean_squared_error(test['cnt'], predictions_random)
```

```python
predictions_random_train = regressor.predict(train[columns])
mean_squared_error(train['cnt'], predictions_random_train)
```

The overfitting is diminished.<br>
The result is less than 1/9th of the  Logistic Regression's one.

The dates in the 'dteday' colummn are strings. <br>
Let's convert them into **datetime** types.

```python
import datetime
def convert(s):
    return datetime.datetime.strptime(s, '%Y-%m-%d')
bike_rentals['dteday'] = [convert(x) for x in bike_rentals['dteday']]
```

```python
fig = plt.figure(figsize=(8,4))
rentals_by_day = bike_rentals.groupby("dteday")["cnt"].sum()
ax = fig.add_subplot(111)
ax.plot(rentals_by_day.index, rentals_by_day.values)
# Turn off tick labels
ax.set_yticklabels([])
ax.set_xticklabels([])
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.axis('off')
ax.set_title("Rentals by Day")
```

The rentals are growing in the period of consideration.

This short analysis was perfect for something that we could name **Logistic Regression versus Decision Tree**.<br>
Decision Tree is a great technique, there are so many cases when it can offer results not even comparable<br>
to the Logistic Regression's ones.<br>
But we always need to keep in mind the **overfitting** issue. We can limit it with a correct use of parameters<br>
or through an ensemble method like **Random Forest**.
